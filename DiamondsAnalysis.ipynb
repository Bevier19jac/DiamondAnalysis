```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df = pd.read_csv(url, index_col=0)

# Display the first few rows to verify the data
df.head()
```

```python
df.tail(7)
```

```python
df.describe()
```

```python
df.describe(include=['object'])
```

```python
print("Unique values in 'cut':", df['cut'].unique())
print("Unique values in 'color':", df['color'].unique())
print("Unique values in 'clarity':", df['clarity'].unique())
```

```python
df.hist(column=['depth', 'table', 'price', 'x', 'y', 'z'], bins=20, figsize=(15, 10))
plt.show()
```

```python
# Bar plot for 'cut'
df['cut'].value_counts().plot(kind='bar', title='Cut')
plt.show()

# Bar plot for 'color'
df['color'].value_counts().plot(kind='bar', title='Color')
plt.show()

# Bar plot for 'clarity'
df['clarity'].value_counts().plot(kind='bar', title='Clarity')
plt.show()
```
